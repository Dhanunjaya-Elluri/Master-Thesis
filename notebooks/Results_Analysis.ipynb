```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

plt.rcParams["figure.figsize"] = (14, 6)

DATA_DIR = "../data/"
```

```python
etth1_results_df = pd.read_csv(DATA_DIR + 'results/ETTh1/result_df.csv')
```

```python
# Calculating the 95% quantile of the "Character Distance"
quantile_95 = etth1_results_df['Character Distance'].quantile(0.95)

# Creating a plot
plt.figure(figsize=(10, 6))
plt.hist(etth1_results_df['Character Distance'], bins=30, color='skyblue', edgecolor='black', alpha=0.7)
plt.axvline(quantile_95, color='red', linestyle='dashed', linewidth=2)
plt.title('Histogram of Character Distance with 95% Quantile')
plt.xlabel('Character Distance')
plt.ylabel('Frequency')
plt.text(quantile_95, plt.ylim()[1]*0.9, f'95% Quantile\n({quantile_95:.2f})', color='red', ha='right')
plt.grid(True)
plt.show()
```

```python
within_95_quantile = etth1_results_df['Character Distance'] <= quantile_95
beyond_95_quantile = etth1_results_df['Character Distance'] > quantile_95

# Creating a plot with different colors for the two groups
plt.figure(figsize=(10, 6))

# Data within the 95% quantile
plt.hist(etth1_results_df['Character Distance'][within_95_quantile], bins=30, color='green', edgecolor='black', alpha=0.7, label='Within 95% Quantile')

# Data beyond the 95% quantile
plt.hist(etth1_results_df['Character Distance'][beyond_95_quantile], bins=30, color='red', edgecolor='black', alpha=0.7, label='Beyond 95% Quantile')

plt.axvline(quantile_95, color='red', linestyle='dashed', linewidth=2)
plt.title('Histogram of Character Distance with Different Colors for 95% Quantile')
plt.xlabel('Character Distance')
plt.ylabel('Frequency')
plt.text(quantile_95, plt.ylim()[1]*0.9, f'95% Quantile\n({quantile_95:.2f})', color='red', ha='right')
plt.legend()
plt.grid(True)
plt.show()
```

```python
sns.histplot(etth1_results_df['Character Distance'][within_95_quantile], color='green', kde=False, binwidth=0.5, label='Within 95% Quantile')

# Plotting the data beyond the 95% quantile
sns.histplot(etth1_results_df['Character Distance'][beyond_95_quantile], color='red', kde=False, binwidth=0.5, label='Beyond 95% Quantile')
plt.axvline(quantile_95, color='red', linestyle='dashed', linewidth=1)
plt.title('Histogram of Character Distance with Different Colors for 95% Quantile')
plt.xlabel('Character Distance')
plt.ylabel('Frequency')
plt.xticks(range(int(etth1_results_df['Character Distance'].min()), int(etth1_results_df['Character Distance'].max()) + 1))
plt.text(quantile_95, plt.ylim()[1]*0.9, f'95% Quantile\n({quantile_95:.2f})', color='red', ha='right')
plt.legend()
plt.grid(True)
plt.show()
```
